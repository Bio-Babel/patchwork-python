# Getting Started

`patchwork` (Python) is a port of the R `patchwork` package. It extends `ggplot2_py` so that multiple plots can be composed into a single figure via arithmetic operators (`+`, `-`, `|`, `/`, `&`, `*`).

In this tutorial we'll work through the basics of using patchwork. By the end you'll have a solid understanding of the API and be ready to dive into the more advanced guides.

> **Operator precedence note**: Python's `|` and `&` bind looser than `/` and `*`, so `p1 | p2 / p3` parses as `p1 | (p2 / p3)` in Python but as `(p1 | p2) / p3` in R. Always parenthesise when mixing operators.

## Setup

```python
from ggplot2_py import aes, facet_wrap, geom_bar, geom_boxplot, geom_point, ggplot, ggtitle, labs

import patchwork as pw
from patchwork._datasets import mtcars

df = mtcars().reset_index()
df.head()
```

## Example plots

We'll start by creating four example plots to use throughout this tutorial.

```python
p1 = ggplot(df) + geom_point(aes(x='mpg', y='disp')) + ggtitle('Plot 1')

p2 = ggplot(df) + geom_boxplot(aes(x='gear', y='disp', group='gear')) + ggtitle('Plot 2')

p3 = ggplot(df) + geom_point(aes(x='hp', y='wt', colour='mpg')) + ggtitle('Plot 3')

p4 = ggplot(df) + geom_bar(aes(x='gear')) + facet_wrap('~cyl') + ggtitle('Plot 4')
```

## Basic use

The absolute simplest use is the `+` operator, which adds two plots together side-by-side.

```python
p1 + p2
```

When adding plots together, the last added plot becomes the "active" one — any subsequent addition of ggplot2 layers, labels, or themes goes to it.

```python
(p1 + p2) + labs(subtitle='This will appear in the last plot')
```

## Controlling layout

By default patchwork tries to produce a square grid filled row-major.

```python
p1 + p2 + p3 + p4
```

Use `plot_layout()` to override:

```python
p1 + p2 + p3 + p4 + pw.plot_layout(nrow=3, byrow=False)
```

## Stacking and packing plots

`|` places plots beside each other; `/` stacks them. Unlike the analogous R vignette, **always parenthesize in Python** because `/` binds tighter than `|`.

```python
p1 / p2
```

Patchworks can be nested, so these two operators alone already cover many complex layouts:

```python
p1 | (p2 / p3)
```

## Annotating the composition

Use `plot_annotation()` to add a title, subtitle, caption, and automatic tagging:

```python
(p1 | (p2 / p3)) + pw.plot_annotation(title='The surprising story about mtcars')
```

Patchwork supports auto-tagging with `tag_levels='A'`, `'a'`, `'1'`, `'i'`, `'I'` — the latter two are lowercase/uppercase Roman numerals.

```python
(p1 + p2 + p3) + pw.plot_annotation(tag_levels='I')
```

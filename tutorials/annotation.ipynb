# Adding Annotation and Style

Sometimes you want multiple plots to read as a single composed figure. Titles, subtitles, captions, and auto-tagging make that easy.

```python
from ggplot2_py import (
    aes, element_text, facet_wrap, geom_bar, geom_boxplot, geom_point,
    ggplot, ggtitle, theme,
)

import patchwork as pw
from patchwork._datasets import mtcars

df = mtcars().reset_index()
p1 = ggplot(df) + geom_point(aes(x='mpg', y='disp')) + ggtitle('Plot 1')
p2 = ggplot(df) + geom_boxplot(aes(x='gear', y='disp', group='gear')) + ggtitle('Plot 2')
p3 = ggplot(df) + geom_point(aes(x='hp', y='wt', colour='mpg')) + ggtitle('Plot 3')
p4 = ggplot(df) + geom_bar(aes(x='gear')) + facet_wrap('~cyl') + ggtitle('Plot 4')
```

## Titles, subtitles and captions

Add a `plot_annotation()` to the patchwork — it only decorates the top-level composition.

```python
patchwork_obj = (p1 + p2) / p3
patchwork_obj + pw.plot_annotation(
    title='The surprising truth about mtcars',
    subtitle='These 3 plots reveal yet-untold secrets about our beloved dataset',
    caption='Disclaimer: None of these plots are insightful',
)
```

## Tagging

Auto-tagging via `tag_levels`. Valid values:
- `'1'` — Arabic numerals
- `'A'` — uppercase Latin letters
- `'a'` — lowercase Latin letters
- `'I'` — uppercase Roman numerals
- `'i'` — lowercase Roman numerals

```python
patchwork_obj + pw.plot_annotation(tag_levels='A')
```

You can style the tag appearance with `&` (which recurses into every sub-plot's theme):

```python
patchwork_obj + pw.plot_annotation(tag_levels='A') & theme(plot_tag=element_text(size=8))
```

Nested patchworks recurse automatically, but `plot_layout(tag_level='new')` starts a fresh level. `tag_levels` can then be a list of tag families:

```python
patchwork_obj[0] = patchwork_obj[0] + pw.plot_layout(tag_level='new')
patchwork_obj + pw.plot_annotation(tag_levels=['A', '1'])
```

`plot_annotation()` also supports `tag_prefix`, `tag_sep`, and `tag_suffix`:

```python
patchwork_obj + pw.plot_annotation(
    tag_levels=['A', '1'],
    tag_prefix='Fig. ',
    tag_sep='.',
    tag_suffix=':',
)
```

You can also pass a custom sequence as a list — mix it with a standard family as needed:

```python
patchwork_obj + pw.plot_annotation(tag_levels=[['#', '&'], '1'])
```

## Styling the patchwork

Most style is carried by the individual ggplots. For the composition itself (background, title, margins) you have two options:

1. Use `&` with a theme — this applies to every sub-plot **and** the patchwork.
2. Use the `theme=` argument of `plot_annotation()` to style only the outer composition.

Both can be combined.

```python
patchwork_obj + pw.plot_annotation(
    title='The surprising truth about mtcars',
    theme=theme(plot_title=element_text(size=18)),
)
```
